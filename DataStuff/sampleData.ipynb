```python
pokemon_mapping={
    "dynamic":"true",
    "properties":{
        "PokeIndex":{"type":"integer"},
        "name":{"type":"text"},
        "Primary type":{"type":"text"},
        "Secondary type":{"type":"text"},
        "Evolves from":{"type":"text"},
        "Evolves into":{"type":"text"},
        "Notes":{"type":"text"}
    }
}
arena_mapping= {
    "dynamic":"true",
    "properties": {
      "name": {"type": "text"},
      "city": {"type": "text"},
      "region": {"type": "text"},
      "leader": {
        "properties": {
          "first_name": {"type": "text"},
          "last_name": {"type": "text"},
          "age": {"type": "integer"},
          "gender": {"type": "keyword"},
          "pokemon": {
            "type": "nested",
            "properties": {
              "id": {"type": "integer"},
              "name": {"type": "text"},
              "level": {"type": "integer"},
              "hp": {"type": "integer"}
            }
          },
        }
      },
      "badge": {"type": "text"},
      "type": {"type": "keyword"}
    }
}
trainer_mapping={
    "dynamic":"true",
    "properties": {
      "first_name": {"type": "text"},
      "last_name": {"type": "text"},
      "age": {"type": "integer"},
      "gender": {"type": "keyword"},
      "pokemon": {
        "type": "nested",
        "properties": {
          "id": {"type": "integer"},
          "name": {"type": "text"},
          "level": {"type": "integer"},
          "hp": {"type": "integer"}
        }
      },
      "badges": {"type": "keyword"}
    }
}
poke_market_mapping= {
    "dynamic":"false",
    "properties": {
      "name": {"type": "text"},
      "location": {"type": "text"},
      "items": {
        "type": "nested",
        "properties": {
          "id": {"type": "integer"},
          "name": {"type": "text"},
          "category": {"type": "keyword"},
          "price": {"type": "integer"},
          "currency": {"type": "keyword"},
          "stock": {"type": "integer"}
        }
      }
    }
}

print("index='pokemon' created:",es.indices.create(index='pokemon', mappings=pokemon_mapping)['acknowledged'])
print("index='arena' created:",es.indices.create(index='arena', mappings=arena_mapping)['acknowledged'])
print("index='trainer' created:",es.indices.create(index='trainer', mappings=trainer_mapping)['acknowledged'])
print("index='poke_market' created:",es.indices.create(index='poke_market', mappings=poke_market_mapping)['acknowledged'])
```

```python
with open('PokemonDataSet.json','r') as f:
    pokemons=json.load(f)
for pokemon in pokemons:
    es.index(index='pokemon',id=pokemon['PokeIndex'] , document=pokemon)
```
